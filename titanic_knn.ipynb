```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler


# Load dataset
iris = load_iris()

data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target


# Separate features and target
X = data.drop('target', axis=1)
y = data['target']


# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)


# 🔍 SHOW TEST DATASET HERE
print("\nFirst 5 rows of TEST DATA:")
print(X_test[:5])

print("\nCorresponding TEST LABELS:")
print(y_test[:5])


# Create KNN model (Manhattan distance)
model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=1)


# Train model
model.fit(X_train, y_train)


# Predict output
predictions = model.predict(X_test)


# Accuracy
accuracy = accuracy_score(y_test, predictions)
print("\nAccuracy:", accuracy)


# Classification report
print("\nClassification Report:")
print(classification_report(y_test, predictions))


# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, predictions))


# View nearest neighbors
distances, indices = model.kneighbors(X_test)

print("\nNearest Neighbor Distances:")
print(distances[:5])

print("\nNearest Neighbor Indices:")
print(indices[:5])
```
